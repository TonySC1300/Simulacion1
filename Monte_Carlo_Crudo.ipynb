```python
import random
import numpy as np
```

```python
def esperado(a, b, n):
    esperanza = 0
    for i in range(n):
        x1 = random.random()
        esperanza += np.sqrt(np.arctan(x1))

    E = esperanza / n  # Valor de la esperanza
    print("El valor aproximado de la integral es:", E)

a = 0
b = 1
n = 10000

esperado(a, b, n)
```

```python
import matplotlib.pyplot as plt

# Definir la función f(x) = sqrt(arctan(x))
def f(x):
    return np.sqrt(np.arctan(x))

# Crear valores de x en el intervalo [0,1]
x = np.linspace(0, 1, 100)
y = f(x)

# Valor de la recta horizontal
y_line = f(1)

# Crear la gráfica
plt.figure(figsize=(5,5))
plt.plot(x, y, label=r'$f(x) = \sqrt{\arctan(x)}$', color='blue')
plt.axvline(x=1, color='red', linestyle='--', label=r'$x=1$')  # Línea vertical
plt.axhline(y=y_line, color='red', linestyle='--', label=rf'$y=\sqrt{{\arctan(1)}}$')  # Línea horizontal

# Configurar límites y etiquetas
plt.xlim(0, 1.2)
plt.ylim(0, 1)

# Rellenar la zona verde debajo de la función en [0,1]
plt.fill_between(x, y, color='green', alpha=0.3)

# Mostrar la gráfica
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función a integrar
def f(x):
    return np.sqrt(np.arctan(x))

# Parámetros de la simulación
N = 10000  # Número de puntos
x_random = np.random.uniform(0, 1, N)  # Generar N puntos aleatorios entre 0 y 1
y_values = f(x_random)  # Evaluar la función en esos puntos

# Calcular la integral aproximada
integral_approx = np.mean(y_values)

# Número de puntos en cada iteración
N_values = np.arange(1, 10001)
integral_approximations = [np.mean(f(np.random.uniform(0, 1, N))) for N in N_values]

# Graficar la convergencia de la integral
plt.plot(N_values, integral_approximations, label='Aproximación de la integral')
plt.axhline(y=integral_approx, color='r', linestyle='--', label='Valor real aproximado')
plt.xlabel('Número de puntos (N)')
plt.ylabel('Valor de la integral')
plt.title('Convergencia de la integral mediante el Método de Montecarlo')
plt.legend()
plt.show()
```
