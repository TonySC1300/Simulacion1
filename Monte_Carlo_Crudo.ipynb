```python
import numpy as np
import random
import matplotlib.pyplot as plt

def esperado_convergencia(a, b, n):
    esperanza = 0
    valores = []  # Lista para almacenar los valores parciales de la esperanza

    for i in range(1, n+1):
        x1 = random.random()
        esperanza += np.sqrt(np.arctan(x1))
        valores.append(esperanza / i)  # Guardar el valor promedio hasta esta iteración

    return valores

# Parámetros
a = 0
b = 1
n = 10000

# Obtener valores de convergencia
valores_convergencia = esperado_convergencia(a, b, n)

# Graficar la convergencia
plt.figure(figsize=(10, 5))
plt.plot(range(1, n+1), valores_convergencia, label="Estimación de la integral", color="b")
plt.axhline(y=valores_convergencia[-1], color="r", linestyle="--", label=f"Valor final: {valores_convergencia[-1]:.5f}")
plt.xlabel("Número de iteraciones")
plt.ylabel("Valor estimado de la integral")
plt.title("Convergencia de la estimación de la integral")
plt.legend()
plt.grid()
plt.show()
```
