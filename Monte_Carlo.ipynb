```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Definir la función f(x) = sqrt(arctan(x))
def f(x):
    return np.sqrt(np.arctan(x))
```

```python
# Crear valores de x en el intervalo [0,1]
x = np.linspace(0, 1, 100)
y = f(x)
```

```python
# Valor de la recta horizontal
y_line = f(1)
```

```python
# Crear la gráfica
plt.figure(figsize=(5,5))
plt.plot(x, y, label=r'$f(x) = \sqrt{\arctan(x)}$', color='blue')
plt.axvline(x=1, color='red', linestyle='--', label=r'$x=1$')  # Línea vertical
plt.axhline(y=y_line, color='red', linestyle='--', label=rf'$y=\sqrt{{\arctan(1)}}$')  # Línea horizontal

# Configurar límites y etiquetas
plt.xlim(0, 1.2)
plt.ylim(0, 1)

# Mostrar la gráfica
plt.show()
```

```python
def dardo(n):
    puntos_x_dentro = []
    puntos_y_dentro = []

    m = 0  # Contador de puntos dentro de la región
    for _ in range(n):
        x1 = random.random()
        y2 = random.random()

        if y2 <= np.sqrt(np.arctan(x1)):
            puntos_x_dentro.append(x1)
            puntos_y_dentro.append(y2)
            m += 1

    p = m / n  # Proporción de puntos dentro de la región
    return p, puntos_x_dentro, puntos_y_dentro
```

```python
def media_dardo(n, k):
    valores_p = [dardo(n)[0] for _ in range(k)]
    return sum(valores_p) / k
```

```python
# Parámetros
n_puntos = 500
k_iteraciones = 10000
```

```python
# Calcular la media de p
media_p = media_dardo(n_puntos, k_iteraciones)
print(f"Media de p tras {k_iteraciones} iteraciones: {media_p}")
```

```python
# Generar puntos con la función dardo
p, x_dentro, y_dentro = dardo(n_puntos)

# Crear la gráfica
x = np.linspace(0, 1, 100)
y = np.sqrt(np.arctan(x))
y_line = np.sqrt(np.arctan(1))

plt.figure(figsize=(5,5))
plt.plot(x, y, label=r'$f(x) = \sqrt{\arctan(x)}$', color='blue')
plt.axvline(x=1, color='red', linestyle='--', label=r'$x=1$')
plt.axhline(y=y_line, color='red', linestyle='--', label=rf'$y=\sqrt{{\arctan(1)}}$')

# Agregar los puntos generados
plt.scatter(x_dentro, y_dentro, color='green', s=10, label="Dentro")

# Configurar límites y etiquetas
plt.xlim(0, 1.2)
plt.ylim(0, 1)
plt.fill_between(x, y, color='yellow', alpha=0.3)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Gráfica con Dardos')
plt.legend()
plt.grid(True)

plt.show()
```
